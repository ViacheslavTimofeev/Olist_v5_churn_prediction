# Информация о размерах и весе товаров (product_measures_olist_public_dataset_.csv)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
product_measures = pd.read_csv(r'C:\Users\Вячеслав\Documents\olist_churn\olist-ml-churn-prediction\data\raw\product_measures_olist_public_dataset_.csv')
```

```python
display(product_measures.head())
print(product_measures.shape)
```

```python
print(product_measures.isna().sum())
```

```python
numerical_features = product_measures.drop('product_id', axis=1).columns
```

```python
n = len(numerical_features)

# Настройка размера фигуры (важно для плотности)
fig, axes = plt.subplots(1, n, figsize=(4 * n, 3))  # один ряд, n колонок

for i, col in enumerate(numerical_features):
    sns.histplot(data=product_measures, x=col, ax=axes[i], kde=False)
    axes[i].set_title(col)

plt.tight_layout()
plt.show()
```

## Выводы из таблицы product_measures:
1. Почти все значения product_weight лежат в диапазоне от 0 до 10000, но также видно, что на 30000 есть всплеск. Это может говорить о наличии фильтра веса на сайте вплоть до 30000, а затем фильтр становится '>30000'.
2. Основная масса товаров варьируется от 18 до 60 сантиметров в длину, от 2 до 60 сантиметров в высоту и от 10 до 50 сантиметров в ширину.
